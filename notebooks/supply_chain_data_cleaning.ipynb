# 1. Import libraries 

```python
import pandas as pd
import numpy as np
```

# 2.Load dataset


```python
df = pd.read_csv(r"C:\Users\User\Desktop\DSI\Abschluss Projekt\supply chain big data\DataSet\DataCoSupplyChainDataset.csv",
    encoding="latin1")
```

# 3. Initial data inspection

```python
# Überblick verschaffen
print("Shape:", df.shape) 
df.info()
display(df.head())
display(df.isnull().sum().sort_values(ascending=False))
```

# 4. Drop irrelevant columns

```python
# Unnötige Spalten entfernen (Diese Spalten sind nicht relevant für Analyse)

columns_to_drop = [
    "Customer Email",
    "Customer Password",
    "Customer Fname",
    "Customer Lname",
    "Product Image"
]

df = df.drop(columns=columns_to_drop, errors="ignore")
```

# 5. Fix data types

```python
# Date

df["order date (DateOrders)"] = pd.to_datetime(df["order date (DateOrders)"], errors="coerce").dt.normalize()
df["shipping date (DateOrders)"] = pd.to_datetime(df["shipping date (DateOrders)"], errors="coerce").dt.normalize()

# Numerische Spalten

df["Sales"] = pd.to_numeric(df["Sales"], errors="coerce")
df["Order Profit Per Order"] = pd.to_numeric(df["Order Profit Per Order"], errors="coerce")


#Check
print(df["order date (DateOrders)"].dtype)
print(df["shipping date (DateOrders)"].dtype)
```

# 6. Check duplicates

```python
#Duplikate prüfen

duplicates = df.duplicated().sum()
print("Number of duplicate rows:", duplicates)
```

# 7. Handle missing values

```python
# Drop columns with very high missingness or low analytical relevance
df = df.drop(columns=["Product Description", "Order Zipcode"], errors="ignore")


# Drop rows with missing customer zipcode because they are required for location-based analysis
df = df.dropna(subset=["Customer Zipcode"])
```

# 8.Final Validation

```python
print("Final shape:", df.shape)
print("Remaining missing values:", df.isnull().sum().sum())
print("Duplicate rows:", df.duplicated().sum())
display(df.head())
```

# 9. Export cleaned dataset

```python
#Exportiere

df.to_csv(
    r"C:\Users\User\Desktop\DSI\Abschluss Projekt\supply chain big data\DataSet\\Cleaned\cleaned_supply_chain_final.csv",
    index=False,
    encoding="utf-8"
)
```
